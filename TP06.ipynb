# Import library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Muat dataset 

```python
df = pd.read_csv("Placement_Data_Full_Class.csv")

# Menampilkan 5 baris pertama
print("--- 5 Baris Pertama Data ---")
print(df.head())

# Melihat informasi umum dataset
print("\n--- Informasi Dataset ---")
print(df.info())

# Menampilkan jumlah data dan kolom
print(f"\nJumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")
```

```python
print(df.columns)
```

# Visualisasi Distribusi

```python
colors = ['#84D2C5', '#F19CBB']  # Placed = toska, Not Placed = pink

counts = df['status'].value_counts()
labels = counts.index
values = counts.values
total = sum(values)

fig, ax = plt.subplots(figsize=(8,6))
bars = ax.bar(
    labels, values, 
    color=colors, 
    edgecolor='black', 
    linewidth=2,
    zorder=3
)

for bar in bars:
    
    ax.bar(
        bar.get_x() + 0.03, 
        bar.get_height() - 0.5, 
        bar.get_width(), 
        color='gray', 
        alpha=0.2, 
        zorder=0
    )

for i, bar in enumerate(bars):
    height = bar.get_height()
    percent = (height / total) * 100
    ax.text(
        bar.get_x() + bar.get_width()/2, 
        height + 2, 
        f'{int(height)}\n({percent:.1f}%)',
        ha='center', va='bottom', fontsize=11, fontweight='bold', color='#333'
    )

ax.set_facecolor('#fdfdfd')
fig.patch.set_facecolor('#f6f7fb')
ax.grid(axis='y', linestyle='--', alpha=0.6, zorder=0)
ax.set_title(
    'Distribusi Status Penempatan Kerja Mahasiswa',
    fontsize=16, fontweight='bold', pad=20
)
ax.set_xlabel('Status (Placed / Not Placed)', fontsize=12, labelpad=10)
ax.set_ylabel('Jumlah Mahasiswa', fontsize=12, labelpad=10)
sns.despine(top=True, right=True)
plt.tight_layout()
plt.show()
```
